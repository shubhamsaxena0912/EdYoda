### Data 1

#### 1. Load the data 

```python
house_rental = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/house_rental_data.csv.txt',index_col='Unnamed: 0')
house_rental.head()
```

#### 2. Find all houses which has more than 2 bedroom and price less than 50k

```python
df = house_rental[(house_rental['Bedroom'] > 2) & (house_rental['Price'] < 50000)]
df.head()
```

#### 3. Find all houses which has more than 3 bedroom and price less than 50k. Update their price with 60k

```python
#Find the houses
df = house_rental[(house_rental['Bedroom'] > 3) & (house_rental['Price'] < 50000)]
df.head()
```

```python
#Update the price
ls = []
updated_house = 0 #This holds record of updated houses

for i in range(len(house_rental)):
    if (house_rental.iloc[i]['Bedroom'] > 3) and (house_rental.iloc[i]['Price'] < 50000):
        ls.append(60000)
        updated_house += 1
    else:
        ls.append(house_rental.iloc[i]['Price'])
        
house_rental['Price'] = ls
```

#### 4. Find no of house who's price is updated in previous question.

```python
print(updated_house)
```

### Data 2 - 'Health care data'

#### Load rental data with first column as index column

```python
hc_df = pd.read_excel('HealthCareData.xlsx')
hc_df.head()
```

#### 1. Display all the information respective to gender

```python
hc_df['Gender'].describe()
```

#### 2. Find all the male who is consuming alcohal more than 5 years and has diabetes result positive.

```python
df = hc_df[(hc_df['Gender'] == 'male') & (hc_df['Duration of alcohol consumption(years)'] > 5) & (hc_df['Diabetes Result'] == 'YES')]
df.head()
```

#### 3.  Find where more people consume alcohol in rural or urban?

```python
rural_cons = hc_df[hc_df['Place(location where the patient lives)'] == 'rural']['Quantity of alcohol consumption (quarters/day)'].values.sum()
urban_cons = hc_df[hc_df['Place(location where the patient lives)'] == 'urban']['Quantity of alcohol consumption (quarters/day)'].values.sum()

if rural_cons > urban_cons:
    print('Rural consumption is more')
    
else:
    print('Urban consumption is more')
```

#### 4. Find average age of male and female who consume alcohol.

```python
# Find details of person who consume alcohol
df = hc_df[hc_df['Quantity of alcohol consumption (quarters/day)'] > 0]

df.groupby('Gender')['Age'].mean()
```

#### 7. Find is there any missing value of age in given data.

```python
hc_df['Age'].isnull().sum()
```

#### 8. If gender is male and age is less than 35 then change diabetes result to NO .

```python
hc_df['Diabetes Result'] = hc_df.apply(lambda x: 'NO' if (x['Gender'] == 'male') and (x['Age'] < 35) else x['Diabetes Result'], axis=1)
```

#### 9. Find all record where Place is missing.

```python
hc_df[hc_df['Place(location where the patient lives)'].isnull()]
```

#### 10. Change place to urban if gender is female else rural.

```python
def fill_place(x):
    if 'female' in x['Gender']:
        return 'urban'
    else:
        return 'rural'
```

```python
#Updation of place
hc_df['Place(location where the patient lives)'] = hc_df.apply(fill_place, axis=1)
```
